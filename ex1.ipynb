```python
def funcMemo(n, memo):
  if n == 0 or n == 1:
    return n
  if not n in memo:
    memo[n] = funcMemo(n-1, memo) + funcMemo(n-2, memo)
  return memo[n]
```

```python
import timeit

def func(n):
  if n == 0 or n == 1:
    return n
  else:
    return func(n-1) + func(n-2)


timeforfunc = []
timeforfuncMemo = []
for i in range(0, 36):
  timeforfunc.append(timeit.timeit(lambda: func(i), setup="from __main__ import func", number=5))
  timeforfuncMemo.append(timeit.timeit(lambda: funcMemo(i,{}), setup="from __main__ import funcMemo", number=5))

print("Time for func: ", timeforfunc)
print("Time for funcMemo: ", timeforfuncMemo)

```

```python
#plot timeforfunc and timeforfuncMemo
import matplotlib.pyplot as plt

integers = [i for i in range(0, 36)]
plt.plot(integers, timeforfunc, label="func")
plt.plot(integers, timeforfuncMemo, label="funcMemo")
plt.legend()
plt.xlabel("Number of integers")
plt.ylabel("Time(s)")
plt.show()
```
